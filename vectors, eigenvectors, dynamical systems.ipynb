```python
%matplotlib inline
from itertools import product as cproduct
import matplotlib.pyplot as plt; plt.style.use("ggplot")
import numpy as np

l = np.array([
        [1/np.sqrt(2), 1/np.sqrt(2)],
        [1/np.sqrt(2), -1/np.sqrt(2)],
    ])
d = np.array([
        [-2, 0],
        [0, 0.5]
    ])

a = l.dot(d).dot(l.T)
print(a)

x_1s, x_2s = np.meshgrid(
    np.linspace(-1, 1, 10, endpoint=True),
    np.linspace(-1, 1, 10, endpoint=True))
f_1s = np.nan * np.zeros(x_1s.shape)
f_2s = np.nan * np.zeros(x_1s.shape)

for i, j in cproduct(range(x_1s.shape[0]), range(x_1s.shape[1])):
    
    x = np.array([x_1s[i, j], x_2s[i, j]])
    f_1s[i, j], f_2s[i, j] = a.dot(x)
    
fig, ax = plt.subplots(1, 1, figsize=(7, 5))

ax.quiver(x_1s, x_2s, f_1s, f_2s, pivot='mid')

ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
```

$
\dot{x_1} = -x_2 + x_1(1 - x_1^2 - x_2^2) \\
\dot{x_2} = x_1 + x_2(1 - x_1^2 - x_2^2) \\
$

```python
%matplotlib inline
from itertools import product as cproduct
import matplotlib.pyplot as plt; plt.style.use('ggplot')
import numpy as np

x_1s, x_2s = np.meshgrid(
    np.linspace(-1.5, 1.5, 15, endpoint=True),
    np.linspace(-1.5, 1.5, 15, endpoint=True))
f_1s = np.nan * np.zeros(x_1s.shape)
f_2s = np.nan * np.zeros(x_1s.shape)

for i, j in cproduct(range(x_1s.shape[0]), range(x_1s.shape[1])):
    
    x = np.array([x_1s[i, j], x_2s[i, j]])
    f_1s[i, j] = -x[1] + x[0]*(1 - x[0]**2 - x[1]**2)
    f_2s[i, j] = x[0] + x[1]*(1 - x[0]**2 - x[1]**2)
    
fig, ax = plt.subplots(1, 1, figsize=(7, 5))

ax.quiver(x_1s, x_2s, f_1s, f_2s, pivot='mid')

ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
```
